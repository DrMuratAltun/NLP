```python
def StyleCloud(file_path='challenge.txt',output_name='stylecloud.png',icon_name='fas fa-coffee'):
    sc.gen_stylecloud(file_path=file_path,
                      icon_name=icon_name,
                      output_name=output_name)
    from PIL import Image
    import matplotlib.pyplot as plt
    im=Image.open(output_name)
    plt.figure(figsize=(10,10))
    plt.imshow(im)
```
