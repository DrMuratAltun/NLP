```python
def metin_temizleme(df_text):
  df_text=df_text.str.lower()
  df_text=df_text.str.replace('[^\w\s]','')
  df_text=df_text.str.replace('[\n,\d+,\r]','')
  return df_text
```

```python
def wc_image(data,bgcolor,mask_image_path):
    plt.figure(figsize=(10,10))
    mask_arry=np.array(Image.open(mask_image_path))
    word_cloud=WordCloud(background_color=bgcolor,stopwords=STOPWORDS, mask=mask_arry)
    word_cloud.generate(' '.join(data))
    plt.imshow(word_cloud)
    plt.axis('off')
```

```python
X=df['Text']
y=df['Class']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from textblob import TextBlob
from nltk.stem import PorterStemmer
pr = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def lemmafn(text):
    words=TextBlob(text).words
    return [pr.stem(word) for word in words]
```

```python
vect=CountVectorizer(stop_words='english',ngram_range=(1,2), analyzer=lemmafn)
```

```python
X=vect.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
nb=MultinomialNB()
```

```python
model=nb.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
accuracy_score(pred,y_test)
```
