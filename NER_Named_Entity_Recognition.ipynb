```python
import spacy
from spacy import displacy

# Modeli yükle
nlp = spacy.load('en_core_web_sm')
```

```python
#python -m spacy download en_core_web_sm (on anaconda prompt)
```

```python
text1='''
Dr. Murat ALTUN is an artificial intelligence instructor and data scientist. 
He has research and books in this field. He greatly enjoys the Python programming language. He works as a Computer Technologies teacher at the Ministry of National Education. 
He also works as an artificial intelligence expert in Turkey.
'''
```

```python
text1=nlp(text1)
```

```python
for word in text1.ents:
    print(word.text,word.label_)
```

```python
spacy.explain('ORG')
```

```python
spacy.explain('PERSON')
```

```python
spacy.explain('GPE')
```

```python
displacy.render(text1)
```
